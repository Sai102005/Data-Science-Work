```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/TSNE_data.csv')
df
```

```python
df.head()
```

```python
array=df.values
X=array[:,1:]
Y=array[:,0]
```

```python
Y
```

```python
array
```

```python
#TSNE Visualization
data_tsne= TSNE(n_components=2).fit_transform(X)
data_tsne
```

```python
newdf=pd.DataFrame(data_tsne).join(df['diagnosis'])
newdf
```

```python
color_class=df['diagnosis'].to_numpy()
cluster.tsneplot(score=data_tsne,colorlist=color_class,legendpos='upper right',legendanchor=(1.51,1))
plt.show()
```
