```python
# change structure of datasets so that we can compute the similarity score
df=movies_df.pivot(index='userId',columns='movie',values='rating')
df
```

```python
df.shape
```

```python
df.fillna(0,inplace=True)
df
```

```python
# Calculating Cosine Similarlity bw users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1- pairwise_distances(df.values,metric='cosine')
user_sim
```

```python
np.fill_diagonal(user_sim,0)
```

```python
user_sim
```

```python
user_sim=pd.DataFrame(user_sim)
```

```python
user_sim
```

```python
user_sim.index=movies_df.userId.unique()
user_sim.columns=movies_df.userId.unique()
user_sim
```
